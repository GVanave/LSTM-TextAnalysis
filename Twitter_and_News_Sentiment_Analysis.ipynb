```python

import pandas as pd
import matplotlib.pyplot as plt
import nltk
import re
from nltk.corpus import stopwords
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
nltk.download('stopwords')
english_stopwords=stopwords.words("english")
```

```python
print(english_stopwords)
```

```python
data_tweet=pd.read_csv("/content/Tweets.csv")
data_tweet
```

```python
(data_tweet.where(data_tweet["sentiment"]=="neutral")).shape
```

```python
dataframe=data_tweet["text"]
print(dataframe
)
```

```python
ps=PorterStemmer()
```

```python

re.sub('[^a-zA-Z]', ' ',dataframe[1]).lower()
```

```python
data=[]
for i in dataframe:
  review=(re.sub('[^a-zA-Z]', ' ',str(i)).lower())
  review=review.split()
  review=[word for word in review if word not in english_stopwords]
  
  review=" ".join(review)
  data.append(review)


  
```

```python
from textblob import TextBlob
```

```python
positive=0
negative=0
neutral=0
for i in data:
  polarity=TextBlob(i).sentiment[0]
  if polarity > 0:
    positive+=1
  elif polarity==0.0:
    neutral+=1
  elif polarity <0:
    negative+=1
  
```

```python
print(positive, negative,neutral)
```

```python
print(f"Percentage of tweets are Positive {(positive/len(data_tweet)*100)}" + "\n" + f"Percentage of tweets are Negative {(negative/len(data_tweet)*100)} ")
```

```python
print(f"percentage tweets arr Neural {100-((positive/len(data_tweet)*100)+(negative/len(data_tweet)*100))}")
```
